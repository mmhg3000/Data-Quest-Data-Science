# The Gender Gap In College Degrees

## Introduction

The goal of this project is to investigate a potential gender gap in the college degrees in 17 majors in the United States.
This shall be achieved via plots for which the **[chart-junk](https://en.wikipedia.org/wiki/Chartjunk)** shall be reduced and the **[data-ink ratio](https://infovis-wiki.net/wiki/Data-Ink_Ratio)** shall be maximized.


## Data

The date was obtained from the The Department of Education Statistics of United States, which is published annually. The report containing data between years 1970 and 2011 was cleaned by Randal Olson, a data scientist at University of Pennsylvania. Randal Olson published cleaned data [here](http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv).


```python
# import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# use the inline magic
%matplotlib inline
```

### Data Exploration

```python
# read the csv file
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
women_degrees.head()
```

### Conclusion
- There are 42 entries for 17 majors in total
- The dtype of Year is int64, for the rest it is float63
- For each year and each major, the percentage of women is listed
- Therefore, the data of Men can be obtained by subtracting the data of Women from 100
- There are no null or missing values

Because there are seventeen degrees, for which line charts are needed for, I will use a subplot grid layout of 6 rows by 3 columns.

I will then group the degrees into
- STEM (Psychology, Biology, Math and Statistics, Physical Sciences, Computer Science, Engineering)
- Liberal Arts (Foreign Languages, English, Communications and Journalism, Art and Performance, Social Sciences and History)
- Other (Health Professions, Public Administration, Education, Agriculture,Business, Architecture)

All three of these lists will be ordered in descending order by the percentage of degrees awarded to women.

Throughout this project, following coloring scheme will be applied, which is color-blind-friendly (refer to following coloring scheme: [Color Blind 10](https://tableaufriction.blogspot.com/2012/11/finally-you-can-use-tableau-data-colors.html)):
- dark blue -> men
- orange -> women

```python
# the rgb values for dark blue and orange
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

# the STEM (Science, Technology, Engineering, Mathematics) categories 
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']

# liberal arts (history, literature, writing, philosophy, sociology, psychology, creative arts, ...) categories
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']

# other categories
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

# creating fig object
fig = plt.figure(figsize=(16, 20))

# plots for STEM
for sp in range(0,6):
    ax = fig.add_subplot(6,3,(sp*3)+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    #ax.spines["right"].set_visible(False)    
    #ax.spines["left"].set_visible(False)
    #ax.spines["top"].set_visible(False)    
    #ax.spines["bottom"].set_visible(False)
    
    # replace above 4 lines by:
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100]) # only display 0 and 100 as y-axis labels
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off') # removing x-axis labels for all plots
    
    # add horizontal line at y = 50 and with transparency = 0.3
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    if sp == 0:
        ax.text(2002, 8, 'Men')
        ax.text(1997, 87, 'Women')
    elif sp == 5:
        ax.text(2002, 87, 'Men')
        ax.text(2000, 8, 'Women')
        ax.tick_params(labelbottom='on') # reinserting x-axis labels for the bottom-most plot

# plots for liberal arts
for sp in range(0,5):
    ax = fig.add_subplot(6,3,(sp*3)+2)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100]) # only display 0 and 100 as y-axis labels
    ax.set_title(lib_arts_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off') # removing x-axis labels for all plots
    
    # add horizontal line at y = 50 and with transparency = 0.3
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    if sp == 0:
        ax.text(2002, 18, 'Men')
        ax.text(1997, 80, 'Women')
    elif sp == 4:
        ax.tick_params(labelbottom='on') # reinserting x-axis labels for the bottom-most plot

# plots for others
for sp in range(0,6):
    ax = fig.add_subplot(6,3,(sp*3)+3)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100]) # only display 0 and 100 as y-axis labels
    ax.set_title(other_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off') # removing x-axis labels for all plots
    
    # add horizontal line at y = 50 and with transparency = 0.3
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    if sp == 0:
        ax.text(2002, 3, 'Men')
        ax.text(1997, 92, 'Women')
    elif sp == 5:
        ax.text(2002, 67, 'Men')
        ax.text(2000, 28, 'Women')
        ax.tick_params(labelbottom='on') # reinserting x-axis labels for the bottom-most plot

# exporting the constructed plots to an image file
plt.savefig('gender_degrees.png')

# viewing the constructed plots
plt.show()
```
